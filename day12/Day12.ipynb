```python
# we definitely have to go with Djikstra's
```

```python
distance = np.full(hn.shape, np.inf)
distance[:5]
```

```python
distance[20][0] = 0
```

```python
Q = []
for i in range(hn.shape[0]):
    for j in range(hn.shape[1]):
        Q.append((i,j))

Q[:5]
```

```python
previous = {x: None for x in Q}
```

```python
neighbors = {}
for vertex in Q:
    neighbors[vertex] = []
    
    # up neighbor
    if vertex[0] > 0 and hv[vertex[0]-1][vertex[1]] - hv[vertex[0]][vertex[1]] <= 1:
        neighbors[vertex].append((vertex[0]-1, vertex[1]))
    
    # down neighbor
    if vertex[0] < 40 and hv[vertex[0]+1][vertex[1]] - hv[vertex[0]][vertex[1]] <= 1:
        neighbors[vertex].append((vertex[0]+1, vertex[1]))
    
    # left neighbor
    if vertex[1] > 0 and hv[vertex[0]][vertex[1]-1] - hv[vertex[0]][vertex[1]] <= 1:
        neighbors[vertex].append((vertex[0], vertex[1]-1))
    
    # right neighbor
    if vertex[1] < 66 and hv[vertex[0]][vertex[1]+1] - hv[vertex[0]][vertex[1]] <= 1:
        neighbors[vertex].append((vertex[0], vertex[1]+1))

neighbors[(0,0)]
```

```python
neighbors[(20,0)]
```

```python
# Djikstra's

while len(Q) > 0:
    q_distances = {k: distance[k] for k in Q}
    min_distance = np.inf
    idx_min = (-1,-1)
    for k in q_distances:
        if q_distances[k] < min_distance:
            min_distance = q_distances[k]
            idx_min = k
    
    Q.remove(idx_min)
    
    for neighbor in neighbors[(idx_min[0], idx_min[1])]:
        alt = distance[idx_min[0]][idx_min[1]] + 1
        if alt < distance[neighbor[0]][neighbor[1]]:
            distance[neighbor[0]][neighbor[1]] = alt
            previous[(neighbor[0], neighbor[1])] = (idx_min[0], idx_min[1])
```

```python
# part 2 -- find the lowest distance from among the 'a' squares == 0

a_squares = []

for i in range(hv.shape[0]):
    for j in range(hv.shape[1]):
        if hv[i][j] == 0:
            a_squares.append((i,j))

a_squares[:5]
```

```python
len(a_squares)
```

```python
a_distances = []

for asq in a_squares:
    distance = np.full(hn.shape, np.inf)
    distance[asq] = 0
    
    Q = []
    for i in range(hn.shape[0]):
        for j in range(hn.shape[1]):
            Q.append((i,j))
    
    previous = {x: None for x in Q}
    
    while len(Q) > 0:
        q_distances = {k: distance[k] for k in Q}
        min_distance = np.inf
        idx_min = (-1,-1)
        for k in q_distances:
            if q_distances[k] < min_distance:
                min_distance = q_distances[k]
                idx_min = k

        Q.remove(idx_min)

        for neighbor in neighbors[(idx_min[0], idx_min[1])]:
            alt = distance[idx_min[0]][idx_min[1]] + 1
            if alt < distance[neighbor[0]][neighbor[1]]:
                distance[neighbor[0]][neighbor[1]] = alt
                previous[(neighbor[0], neighbor[1])] = (idx_min[0], idx_min[1])
    
    a_distances.append(distance[(20,43)])

min(a_distances)
```
