```python
# read the forest grid into a data frame
forest = pd.read_fwf("input.txt", widths=[1]*99, header=None)
forest.head()
```

```python
# check visibility from left
visible_from_left = (forest.cummax(axis='columns') > forest.cummax(axis='columns').shift(1, axis='columns').replace(np.nan, 0))
visible_from_left.head()
```

```python
# check visibility from top
visible_from_top = (forest.cummax(axis='rows') > forest.cummax(axis='rows').shift(1, axis='rows').replace(np.nan, 0))
visible_from_top.head()
```

```python
# check visibility from right
# first reverse order of columns

right_forest = forest[list(range(98,-1,-1))]
right_forest.head()
```

```python
visible_from_right = (right_forest.cummax(axis='columns') > right_forest.cummax(axis='columns').shift(1, axis='columns').replace(np.nan, 0))
visible_from_right.head()
```

```python
# check visibility from bottom
# first reverse order of rows

bottom_forest = forest.loc[list(range(98,-1,-1))]
bottom_forest.head()
```

```python
visible_from_bottom = (bottom_forest.cummax(axis='rows') > bottom_forest.cummax(axis='rows').shift(1, axis='rows').replace(np.nan, 0))
visible_from_bottom.head()
```

```python
visible_from_left[0] = True
visible_from_right[98] = True
```

```python
visible_from_top.loc[0] = True
visible_from_top.head()
```

```python
visible_from_bottom.loc[98] = True
visible_from_bottom.head()
```

```python
visible_from_bottom = visible_from_bottom.loc[list(range(99))]
visible_from_bottom.head()
```

```python
visible_from_bottom.tail()
```

```python
visible_from_right = visible_from_right[list(range(99))]
visible_from_right.head()
```

```python
visible = visible_from_left | visible_from_right | visible_from_bottom | visible_from_top
visible
```

```python
forest_grid = forest.to_numpy()
forest_grid[:5]
```

```python
distance_scores = np.zeros(shape=(99,99), dtype=int)

# skip the edges
for i in range(1,98):
    for j in range(1,98):
        # look up
        position = i-1
        up_counter = 1
        while position > 0 and forest_grid[position][j] < forest_grid[i][j]:
            position -= 1
            up_counter += 1
        # look down
        position = i+1
        down_counter = 1
        while position < 98 and forest_grid[position][j] < forest_grid[i][j]:
            position += 1
            down_counter +=1
        # look left
        position = j-1
        left_counter = 1
        while position > 0 and forest_grid[i][position] < forest_grid[i][j]:
            position -= 1
            left_counter += 1
        # look right
        position = j+1
        right_counter = 1
        while position < 98 and forest_grid[i][position] < forest_grid[i][j]:
            position += 1
            right_counter += 1
        
        # update viewing score
        distance_scores[i][j] = up_counter * down_counter * left_counter * right_counter

distance_scores[:5]
```

```python
np.max(distance_scores)
```
