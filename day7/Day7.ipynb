```python
# Advent of Code 2022 - Day 7
# Moideen Kalladi

# open and read input file
with open("input.txt", "r") as f:
    lines = f.read().split("\n")
```

```python
# so once we have the file system in a tree-like data structure, getting the sizes is a simple recursive function
# the tricky part is building up the data structure from the commands
```

```python
class File:
    name = ""
    size = 0
    parent = None
    otype = "file"
   
    def __init__(self, fname, fsize, fparent):
        self.name = fname
        self.size = fsize
        self.parent = fparent

class Directory:
    name = ""
    contents = []
    parent = None
    otype = "dir"
    
    def __init__(self, dname, dparent):
        self.name = dname
        self.parent = dparent
        self.contents = []
    
    def add_obj(self, obj):
        self.contents.append(obj)
    
    def get_dir(self, dname):
        for obj in self.contents:
            if obj.name == dname:
                return obj
        return None
```

```python
root = Directory("root", None)
current_dir = root
ls_flag = False

for line in lines[1:]:
    # print(line)
    if line == "":
        continue
    
    if line[:4] == "$ ls":
        ls_flag = True
        continue
    
    if line[:7] == "$ cd ..":
        ls_flag = False
        current_dir = current_dir.parent
        print(f"current dir is now {current_dir.name}")
        continue
    
    if line[:4] == "$ cd":
        ls_flag = False
        dname = line[5:]
        current_dir = current_dir.get_dir(dname)
        print(f"current dir is now {current_dir.name}")
        continue
    
    if ls_flag is True:
        if line[:3] == "dir":
            dname = line[4:]
            newdir = Directory(dname, current_dir)
            current_dir.add_obj(newdir)
            print(f"{dname} added to {current_dir.name}")
        else:
            lparts = line.split(" ")
            fsize = int(lparts[0])
            fname = lparts[1]
            newfile = File(fname, fsize, current_dir)
            current_dir.add_obj(newfile)
            print(f"{fname} added to {current_dir.name}")
        continue
```

```python
def print_sizes(direc):
    size_sum = 0
    for obj in direc.contents:
        if obj.otype == "file":
            size_sum += obj.size
        else:
            size_sum += print_sizes(obj)
    print(f"{direc.name} {size_sum}")
    return size_sum
```

```python
print_sizes(root)
```
